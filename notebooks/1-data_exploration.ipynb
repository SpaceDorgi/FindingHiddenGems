```python
import json
```

```python
businesses = []
with open('../yelp_dataset/yelp_academic_dataset_business.json', 'r', encoding='utf-8') as f:
    for line in f:
        business = json.loads(line.strip())
        businesses.append(business)
```

```python
businesses[0]
```

```python
businesses[:11]
```

### Get businesses that are restaurants

```python
restaurants = []
for b in businesses:
    if  b['categories'] is not None and ('food' in b['categories'].lower() or 'restaurant' in b['categories'].lower()):
        restaurants.append(b)
len(restaurants)
```

```python
restaurants = {"restaurants": restaurants}
# with open('../yelp_dataset/yelp_restaurants.json', 'w', encoding='utf-8') as f:
#     json.dump(restaurants, f, indent=2)
```

```python
rest_ids = set([r['business_id'] for r in restaurants['restaurants']])
```

```python
category_list = []
for r in restaurants['restaurants']:
    categories = r['categories'].split(', ')
    category_list.extend(categories)
print(len(category_list))
```

```python
len(set(category_list))
```

```python
reviews = []
with open('../yelp_dataset/yelp_academic_dataset_review.json', 'r', encoding='utf-8') as f:
    for line in f:
        review = json.loads(line.strip())
        reviews.append(review)

len(reviews)
```

```python
restaurant_reviews = [r for r in reviews if r['business_id'] in rest_ids]
len(restaurant_reviews)
```
